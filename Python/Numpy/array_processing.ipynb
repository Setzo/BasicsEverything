```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
points = np.arange(-5, 5, 0.01)
```

```python
dx, dy = np.meshgrid(points, points)
```

```python
dx
```

```python
dy
```

```python
z = (np.sin(dx) + np.sin(dy))
z
```

```python
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()
plt.title('Plot of dx, dy -> sin(dx) + sin(dy)')
```

```python
A1 = np.array([1, 2, 3, 4])
A2 = np.array([100, 200, 300, 400])
conditionArray = np.array([True, True, False, False])
```

```python
zip(A1, A2, conditionArray)
```

```python
conditionResultArray = [(A1_value if condition else A2_value) for A1_value, A2_value, condition in zip(A1, A2, conditionArray)]
conditionResultArray
```

```python
conditionResultArrayNp = np.where(conditionArray, A1, A2)
conditionResultArrayNp
```
